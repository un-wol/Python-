print_even ([1, 3, 2, 10, 12, 11, 15])

```python
def print_score(numbers):
    for num in numbers:
        if num % 2 == 0:
            print(num)

print_score([1, 3, 2, 10, 12, 11, 15])
```

__[문제7]__

오즈 (oz) 클래스에서 이름, 나이, 성별을 출력하는 who() 메소드를 추가하고 who()메소드를 이용해 아래 출력 결과를 만들어주세요


출력 결과 ) <br>
이름: 김코, 나이: 30, 코딩레벨: A

```python
class oz:
    def __init__(self, 이름, 나이, 코딩레벨):
        self.이름 = 이름
        self.나이 = 나이
        self.코딩레벨 = 코딩레벨

    def who(self):
        print(f"이름: {self.이름}, 나이: {self.나이}, 코딩레벨: {self.코딩레벨}")

coding = oz("김코", 30, "A")

coding.who()
```
