__[문제3]<br>__

다음 리스트에서 최댓값과 최솟값, 그리고 평균을 출력해주세요.
<br>

```
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
```



````python
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]

최댓값 = max(nums)
최솟값 = min(nums)
평균 = sum(nums) / len(nums)

print("최댓값:", 최댓값)
print("최솟값:", 최솟값)
print("평균:", 평균)
````
