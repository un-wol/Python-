__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
wizard_intelligence = 80

can_cast_spell = wizard_intelligence >= 50

print( can_cast_spell )
```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
warrior_strength = 80

warrior_health = 100

can_strong_attack = warrior_strength >= 60 and warrior_health >= 80

print( can_strong_attack )
```

<br>

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
bandit_agility = 80

bandit_strength = 40

can_dodge_and_attack = bandit_agility >= 70 or bandit_strength >= 50

print( can_dodge_and_attack )
```

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
healer_intelligence = 70

healer_health = 70

can_heal = healer_intelligence >= 60 and healer_health >= 50

print( can_heal )
```
